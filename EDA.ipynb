**Exploratry Data Analysis (EDA)**

```python
ax=sns.countplot(x='Zone',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
bx=sns.countplot(x='State',data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in bx.containers:
    bx.bar_label(bars)
```

```python
cx=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in cx.containers:
    cx.bar_label(bars)
```

```python
dx=sns.countplot(x='Orders',data=df)
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
fx=sns.countplot(x='Age',data=df)
for bars in fx.containers:
    sns.set(rc={'figure.figsize':(15,5)})
    fx.bar_label(bars)
```

```python
Khadija=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.pointplot(x='State',y='Amount',data=Khadija)
```

```python
Phone=df.groupby(['Cust_name'],as_index=False)['Occupation'].sum().sort_values('Occupation',ascending=False)

sns.barplot(x='Cust_name',y='Occupation',data=Phone)
```
